### common libraries

```python
import math

# Calculate square root
print(math.sqrt(16))  # Output: 4.0

# Calculate factorial
print(math.factorial(5))  # Output: 120
```

```python
import random

# Generate a random number between 1 and 10
print(random.randint(1, 10))  # Output: Random number between 1 and 10

# Choose a random element from a list
items = ['apple', 'banana', 'cherry']
print(random.choice(items))  # Output: Random item from the list
```

```python
from datetime import datetime

# Get current date and time
now = datetime.now()
print("Current date and time:", now)

# Format the date
formatted_date = now.strftime("%Y-%m-%d %H:%M:%S")
print("Formatted date:", formatted_date)
```

```python
import numpy as np

#numerical operations and arrays

# Create an array
arr = np.array([1, 2, 3, 4])
print("Array:", arr)

# Perform mathematical operations
print("Sum:", np.sum(arr))  # Output: 10
```

```python
import pandas as pd

#dataframes

# Create a DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35]}
df = pd.DataFrame(data)
print(df)

# Perform basic operations
print("Mean age:", df['Age'].mean())  # Output: Mean age: 30.0
```

## matplotlib

static, animated, and interactive plots

visualization

```python
import matplotlib.pyplot as plt

# Simple plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
print(x)
print(y)
plt.plot(x, y)  # Plot y = sin(x)
plt.show()  # Display the plot
```

```python
plt.plot(x, y, label="sin(x)", color='blue')
plt.title("Plot of sin(x)")
plt.xlabel("x axis")
plt.ylabel("sin(x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, label="sin(x)")
plt.plot(x, y2, label="cos(x)")
plt.legend()
plt.show()
```
